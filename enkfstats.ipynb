```python
def innovation(a):
    temp = a.variables['obs_type'][:]
    obtype = 37
    t = temp[:] == obtype
    innov = a.variables['observations'][t,0] - a.variables['observations'][t,2]
    return innov

```

```python
def totspread(a):
    sigmavr = 2.0
    N = 50.
    dof = 1./49.
    temp = a.variables['obs_type'][:]
    obtype = 37
    t = temp[:] == obtype
    temp2 = a.variables['observations'][t,:]
    avg = np.average(temp2[:,0])
#    print avg
    test = np.average(dof*(np.sum(((temp2[:,0] - avg)**2))))
#    print test
    return np.sqrt(sigmavr**2 + np.average(dof*(test)))
    
```

```python
innov = np.zeros(len(glob.glob('/Volumes/temp/obs_epoc/obs*.nc')))
for k,i in enumerate(glob.glob('/Volumes/temp/obs_epoc/obs*.nc')):
    a = nc.Dataset(i,'r')
    temp = innovation(a)
    atevmp = np.average(temp)
    dif = temp - atevmp
    innov[k] = np.sqrt(np.square(np.average(dif)))
    a.close()
    
```

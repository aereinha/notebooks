```python
def pressure(znw,mu,mub,gpttot,theta,qvapor):
    rdgas = 287.
    rvgas = 461.6
    kap = 2./7.
    kapdiv = 1./(1.-kap)
    prefcont = 100000.**(-(kap))
    surfptotal = mu+mub
    pres = gpttot[0:-1,:,:]
    rhodry = gpttot[0:-1,:,:]
    for k in range(len(gpttot)-1):
        levee = znw[k+1]-znw[k]
        rhodry[k,:,:] = -((levee)/(gpttot[k+1,:,:]-gpttot[k,:,:]))*surfptotal
    totpres = (theta*prefcont*rhodry*rdgas*(1.+(rvgas/rdgas)*qvapor))**kapdiv
    return totpres
```

```python
def air_density(theta,p):
    rdgas = 287.
    preffer = 100000.
    rvgas = 461.6
    kap = 2./7.
    T = theta * ((preffer/p)**(-kap))
    return T, p/(rdgas*T)
    
```

```python
def calcn0rr(nr3d,qr3d):
    RHOW = 997.
    PI = np.pi
    LAMINR = 1./2800.e-6
    LAMAXR = 1./20.e-6
    
    lamr = (PI * RHOW * nr3d/qr3d)**(1./3.)
    
    #Adjust vars
    lamr[lamr< LAMINR] = LAMINR
    lamr[lamr>LAMAXR] = LAMAXR
    n0rr = nr3d * lamr
    
    return lamr,n0rr
    
```

```python
def calcn0gg(ng3d,qg3d):
    rhog = 900.
    cg = rhog * np.pi/6.
    cons2 = gamma(1.+3.)*cg
    dg = 3.
    lamg = (cons2*ng3d/qg3d)**(1./dg)
    n0gg = ng3d*lamg
    return lamg,n0gg
```

```python
def calcn0rrmy(qnrain,qrain,p,psfc,t):
    RHOW = 997.
    Rd = 287.
    PI = np.pi
    LAMINR = 1./2800.e-6
    LAMAXR = 1./20.e-6
    qnrain = qnrain*((p/psfc)*psfc/(Rd*t))
    de = calcdeMY(p,psfc,t,qrain,qnrain)
    cmr = (np.pi/6.)*1000.
    
    gr34 = gamma(4.)
    gr31 = gamma(1.)
    igr31 = 1./gr31
    cexr9 = cmr*gr34*igr31
    ilamr = (de*qrain*(1./qnrain)*(1./cexr9))**(1./3.)
    lamr = 1./ilamr
    #Adjust vars
   # lamr[lamr< LAMINR] = LAMINR
   # lamr[lamr>LAMAXR] = LAMAXR
    n0rr = qnrain * lamr
    
    return lamr,n0rr
    
```

```python
def calcdeMY(p,psfc,t,qrain,qnrain):
    den = (p/psfc)*p/(287.*t)
    dmx = den*qrain*(1./qnrain)*((np.pi/6)*1000.**(1./3.))
    return dmx
```

```python
def gamma(xx):
    x = xx
    y = x
    cof = [76.18009172947146,-86.50532032941677,24.01409824083091,-1.231739572450155,0.001208650973866179,-0.000005395239384953]
    stp = 2.5066282746310005
    tmp = (x+5.5)
    tmp = (x+0.5)*np.log(tmp)-tmp
    ser = 1.00000000190015
    for j in range(1,5):
        y = y+1.0
        ser = ser+cof[j]/y
    gammadp = tmp+np.log(stp*ser/x)
    gammadp = np.exp(gammadp)
    return gammadp
    
    
```
